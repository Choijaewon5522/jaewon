```python
import csv
information = open('crime.csv')
data = csv.reader(information)

crime = input('범죄 종류: ')

city = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종']
result = [0 for _ in range(8)]
for row in data:
    if row[0] == crime:
        for i in range(8):
            result[i] += int(row[3+i])

min_value = min(result)
for i in range(8):
    result[i] /= min_value
    print(city[i], result[i])

import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family='Malgun Gothic')
size = result
labels = '서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종'
colors = ['yellowgreen', 'gold', 'red', 'lightskyblue', 'lightcoral', 'blue', 'pink', 'orange']

plt.title('도시별 ' + crime + '비율')
plt.pie(size, labels=labels, colors=colors, autopct='%1.1f%%', startangle=90)
plt.axis('equal')
plt.savefig('도시별' + crime +'.png')
plt.show()
```

```python
import csv
information = open('crime.csv')
data = csv.reader(information)

crime = input('범죄 종류: ')

city = ['서울', '부산', '대구', '인천', '광주', '대전', '울산', '세종']
result = [0 for _ in range(8)]
population = [10018537, 3513361, 2487823, 2929195, 1472802, 1518024, 1173568, 215418]
for row in data:
    if row[0] == crime:
        for i in range(8):
            result[i] += int(row[3+i])

for i in range(8):
    result[i] /= population[i]

min_value = min(result)
for i in range(8):
    result[i] /= min_value
    print(city[i], result[i])
    
import matplotlib.pyplot as plt
plt.figure(dpi=300)
plt.rc('font', family='Malgun Gothic')
plt.title('도시별 ' + crime)

plt.bar(range(8), result, width=0.4)
plt.xticks(range(8), city)
plt.xlabel('도시')
plt.ylabel('범죄수/인구수')
plt.legend()
plt.savefig('인구수에 따른 도시별' + crime +'.png')
plt.show()
```
